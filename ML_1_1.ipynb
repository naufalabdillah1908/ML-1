```python
#read data
data = pd.read_csv('arrhythmia.data', sep=',', header = None)
data.columns = cols
data
```

**CARI MISSING VALUE AND REPLACE**

```python
#replace ? to nan
data = data.replace("?", np.nan)
data.isnull().sum()
```

```python
pd.isnull(data).sum().plot()
plt.xlabel('Feature')
plt.ylabel('NaN')
```

```python
pd.isnull(data.iloc[:, 10:15]).sum().plot()
```

```python
#drop j, karena banyak nan
data = data.drop(['j'], axis=1)
```

```python
#fill sisanya pakai modus

for i, s in data.items():
  data.fillna(data[i].mode()[0], inplace = True)
```

```python
#cek jika masih ada yang nan
data.isnull().sum().sum()
```

```python
#cek jika data int atau float semua
for i, s in data.items():
  if not isinstance(data[i][0], int) or not isinstance(data[i][0], float):
    data[i] = data[i].astype(int)
```

**DROP COLLUMN WITH LOW/NO VARIANCE AND LOW INFO GAIN**

```python
#Check if ada column yang variancenya 0, jika ada drop
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold=0)
selector.fit_transform(data)
data
```

```python
#cek info gain/mutual info pada column
from sklearn.feature_selection import mutual_info_classif

x_data = data.iloc[:, 0:278]
y_data = data.iloc[:, 278:279]
mutual_info = mutual_info_classif(x_data, np.ravel(y_data))
mutual_info = pd.Series(mutual_info)
mutual_info.index = x_data.columns
mutual_info.sort_values(ascending=False)
```

```python
#plot bar buat mutual gain
mutual_info.sort_values(ascending=False).plot.bar(figsize=(50,10))
```

```python
#ambil yang >= 0.n

mutual_info = mutual_info[mutual_info >= 0.1]
mutual_info.sort_values(ascending=False).plot.bar(figsize=(10,10))
```

```python
count = len(mutual_info)
#select yang count terbagus
from sklearn.feature_selection import SelectKBest
kbest = SelectKBest(mutual_info_classif, k=count)
kbest.fit(x_data, y_data)
hasil = x_data.columns[kbest.get_support()]
```

```python
x_data = x_data[hasil]
```

```python
#code outlier
def outlier_removal(data):
  new = sorted(data)
  q1, q3 = np.percentile(new,[25,75])
  iqr = q3-q1

  max_o = q3 + (1.5*iqr)
  min_o = q1 - (1.5*iqr)

  for i, s in data.items():
    if s > max_o:
      data[i] = np.nan
    if s < min_o:
      data[i] = np.nan
```

```python
for i, s in x_data.items():
  outlier_removal(s)
```
